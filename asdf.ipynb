```python
def solution(n):
    answer = [n,]
    while True:
        if n % 2 == 0:
            n /= 2
        else:
            n = 3 * n + 1
        answer.append(n)
        if answer[-1] == 1:
            break
    return answer
solution(10)
```

```python
from collections import Counter

def solution(a, b, c, d):
    cnt = Counter([a, b, c, d])           # {눈:개수}
    items = sorted(cnt.items(), key=lambda x: x[1], reverse=True)  # 개수 내림차순

    # (1) 4개 같음: {p:4}
    if items[0][1] == 4:
        p = items[0][0]
        return 1111 * p

    # (2) 3개 같고 1개 다름: {p:3, q:1}
    if items[0][1] == 3:
        p = items[0][0]
        q = items[1][0]
        return (10 * p + q) ** 2

    # (3) 2개씩 같음: {p:2, q:2}
    if items[0][1] == 2 and items[1][1] == 2:
        p = items[0][0]
        q = items[1][0]
        return (p + q) * abs(p - q)

    # (4) 2개 같고 나머지 2개는 서로 다름: {p:2, q:1, r:1}
    if items[0][1] == 2:
        singles = [x for x, k in cnt.items() if k == 1]  # [q, r]
        return singles[0] * singles[1]

    # (5) 모두 다름: {x:1, y:1, z:1, w:1}
    return min(a, b, c, d)
```
