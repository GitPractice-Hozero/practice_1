### 최근접이웃 분류 모델로 훈련 및 평가 해보기

```python
# (데이터 로드 및 정제)
# 1. input data, target, data를 가지고 DataFrame 만들어서 describe() 확인
# 2. 데이터 전처리
#   (1) 입력 - 정답 데이터 분리 (O)
#   (2) 훈련 - 평가 데이터 분리
#   (3) 정규화
# 3. KNN분류모델 학습
# 4. 학습된 모델 평가

import pandas as pd
import numpy as np

df = pd.DataFrame(np.array(iris_data.data), columns=iris_data.feature_names)
df

df
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_label, test_label = train_test_split(iris_data.data ,iris_data.target,test_size=.10,random_state=0,stratify=iris_data.target)
```

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
train_scaled = std_scaler.fit_transform(train_input)
test_scaled = std_scaler.transform(test_input)
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
kn.fit(train_scaled, train_label)
kn.score(train_scaled,train_label)
```
